```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# Load data
df = pd.read_csv('Sample Points.csv')

# Convert deformation columns to numeric, coercing errors
for col in ['Y1517', 'Y1921', 'Y2325']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Period columns mapping
period_cols = {
    '2015-2017': 'Y1517',
    '2019-2021': 'Y1921',
    '2023-2025': 'Y2325'
}

# Compute ECDF and plot
plt.figure()
for period, col in period_cols.items():
    data = np.sort(df[col].dropna().values)
    ecdf = np.arange(1, len(data) + 1) / len(data)
    plt.step(data, ecdf, where='post', label=period)
plt.title('ECDF of Deformation by Monitoring Period')
plt.xlabel('Deformation (mm)')
plt.ylabel('Empirical Cumulative Distribution')
plt.legend()
plt.grid(True)
plt.show()

# Define quantiles
quantiles = [0, 25, 50, 75, 100]
index_labels = [f"{q}%" for q in quantiles]

# Build quantile table
quantile_vals = {
    period: [df[col].quantile(q/100) for q in quantiles]
    for period, col in period_cols.items()
}
quantile_df = pd.DataFrame(quantile_vals, index=index_labels)

# Display the quantile table
display(quantile_df)
```

```python
import pandas as pd
from scipy.stats import wilcoxon

# Load data
df = pd.read_csv('Sample Points.csv')

# Define period-column mapping
pairs = [
    ('2015-2017', '2019-2021', 'Y1517', 'Y1921'),
    ('2015-2017', '2023-2025', 'Y1517', 'Y2325'),
    ('2019-2021', '2023-2025', 'Y1921', 'Y2325'),
]

# Compute Wilcoxon signed-rank test for each paired interval
results = []
for int1, int2, col1, col2 in pairs:
    data = df[[col1, col2]].dropna()
    stat, p = wilcoxon(data[col1], data[col2])
    results.append({
        'Interval 1': int1,
        'Interval 2': int2,
        'Wilcoxon_stat': round(stat, 3),
        'p-value': round(p, 5),
        'Significance (=0.05)': 'Significant' if p < 0.05 else 'Not significant'
    })

# Create results table
res_df = pd.DataFrame(results)
print("Paired Wilcoxon Signed-Rank Test Results:\n")
print(res_df.to_string(index=False))
```
