# Matplotlib Labels and Title

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4, 20])
y = np.array([2, 18])
plt.ylabel("Blake is an old man")
plt.xlabel("Pablo is a young man")


plt.plot(x)
plt.plot(y) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4, 20])
y = np.array([2, 18])
plt.ylabel("Blake is an old man")
plt.xlabel("Pablo is a young man")
plt.title("Kingsley Chika CHUKWU")


plt.plot(x)
plt.plot(y) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4, 20])
y = np.array([2, 18])

font1 = {"family":"arial", "color": "green", "size": 20}
font2 = {"family":"serif", "color": "black", "size": 10}

plt.ylabel("Blake is an old man", fontdict = font2)
plt.xlabel("Pablo is a young man",  fontdict = font2)
plt.title("Kingsley Chika CHUKWU",  fontdict = font1)


plt.plot(x)
plt.plot(y) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4, 20])
y = np.array([2, 18])

font1 = {"family":"arial", "color": "green", "size": 20}
font2 = {"family":"serif", "color": "black", "size": 10}

plt.ylabel("Blake is an old man", fontdict = font2)
plt.xlabel("Pablo is a young man",  fontdict = font2)
plt.title("Kingsley Chika CHUKWU",  fontdict = font1, loc = "right")


plt.plot(x)
plt.plot(y) 
plt.show()
```
