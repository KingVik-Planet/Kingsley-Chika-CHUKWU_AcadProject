```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
Begineers = ["Pablo", "Steph", "Collins", "Blake"]
value = [450, 500, 200, 150]
bar_labels = ["red", "blue", "orange", "green"]
bar_colors = ["tab:red", "tab:blue","tab:orange","tab:green"]
ax.bar(Begineers, value, label = bar_labels , color = bar_colors)
ax.set_ylabel("Begineers Value")
ax.set_xlabel("Begineers Name")
ax.legend(title = "Beg Legend")
ax.set_title("Begineers Weight")
plt.show()
```
