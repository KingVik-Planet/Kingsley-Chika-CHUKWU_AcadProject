```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1) Load data and aggregate
df = pd.read_excel('Data/UNILAK/Rainfall and Temprature.xlsx')
annual = df.groupby('Year').agg({
    'PRECIP': 'sum',      # total annual rainfall
    'TMPMAX': 'mean',     # average annual max temp
    'TMPMIN': 'mean'      # average annual min temp
})
# Select only the years of interest
years = [2014, 2019, 2024]
annual_sel = annual.loc[years]

# 2) Plotting
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

#  Panel A: Bar & dualaxis line
x = np.arange(len(years))
axes[0].bar(x, annual_sel['PRECIP'], width=0.4, label='Annual Rainfall (mm)')
axes[0].set_xticks(x)
axes[0].set_xticklabels(years)
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Rainfall (mm)')
axes[0].set_title('Annual Rainfall and Temperature')
axes[0].grid(axis='y', linestyle='--', linewidth=0.5)

ax2 = axes[0].twinx()
ax2.plot(x, annual_sel['TMPMAX'], marker='o', label='Avg Tmax (C)')
ax2.plot(x, annual_sel['TMPMIN'], marker='s', label='Avg Tmin (C)')
ax2.set_ylabel('Temperature (C)')

# Move legends so they dont overlap
axes[0].legend(loc='upper right')
ax2.legend(loc='upper left')

#  Panel B: Correlation heatmap
corr = annual_sel[['PRECIP','TMPMAX','TMPMIN']].corr()
im = axes[1].imshow(corr, vmin=-1, vmax=1)
axes[1].set_xticks(range(3))
axes[1].set_yticks(range(3))
axes[1].set_xticklabels(corr.columns)
axes[1].set_yticklabels(corr.index)
axes[1].set_title('Correlation Matrix')
# Annotate values
for i in range(3):
    for j in range(3):
        axes[1].text(j, i, f"{corr.iloc[i,j]:.2f}",
                     ha='center', va='center',
                     color='white' if abs(corr.iloc[i,j])>0.5 else 'black')
# Colorbar
fig.colorbar(im, ax=axes[1], fraction=0.046, pad=0.04).set_label('Corr. coeff')

plt.tight_layout()
plt.show()
```
